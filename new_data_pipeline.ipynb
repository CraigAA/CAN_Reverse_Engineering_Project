```python
indicator_annot_path = 'CAN_Reverse_Engineering_Project/logs/6_2012 Subaru Impreza/indicator/6_indicator_annot.log'
df_annot = pd.read_csv(indicator_annot_path, sep=',', header=None, names=['timestamp', 'action'])
df_annot
```

```python
# 1. Clean the CAN DataFrame
# Remove the parentheses from the timestamps and convert them to floats
df_can['timestamp'] = df_can['timestamp'].astype(str).str.replace(r'\(|\)', '', regex=True).astype(float)

# Split the 'raw_frame' column (e.g., '140#0006...') into separate 'id' and 'payload' columns
df_can[['id', 'payload']] = df_can['raw_frame'].str.split('#', expand=True)

# 2. Isolate the target ID
target_id = '152'
df_152 = df_can[df_can['id'] == target_id].copy()

# 3. Create a function to find the closest payload to the annotated timestamp
def get_payload_at_time(df, target_time):
    # Find the CAN messages broadcast in the 1 second after the switch was flipped
    mask = (df['timestamp'] >= target_time) & (df['timestamp'] <= target_time + 1.0)
    if not mask.any():
        return None
    # Return the first payload found
    return df[mask].iloc[0]['payload']

# 4. Cross-reference the annotations with the CAN traffic
results = []
for index, row in df_annot.iterrows():
    action_time = row['timestamp']
    action_name = row['action']
    
    payload = get_payload_at_time(df_152, action_time)
    
    if payload and len(payload) == 16: # Ensure we have a full 8-byte payload
        # Extract Byte 7 (the last 2 characters of the 16-character hex string)
        byte_7_hex = payload[14:16]
        
        # Convert hex to an 8-bit binary string for visualization
        byte_7_bin = bin(int(byte_7_hex, 16))[2:].zfill(8)
        
        results.append({
            'Action': action_name,
            'Time': action_time,
            'Full Payload': payload,
            'Byte 7 (Hex)': byte_7_hex,
            'Byte 7 (Bin)': byte_7_bin
        })

# 5. Display the verification table
df_results = pd.DataFrame(results)
print(f"--- Verification for CAN ID {target_id} ---")
df_results
```
