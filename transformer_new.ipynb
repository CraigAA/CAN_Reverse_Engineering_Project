```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np

class CANClassificationDataset(Dataset):
    def __init__(self, bit_data_list, labels_list, start_indices_list, window_size=200):
        self.window_size = window_size
        self.bit_data_list = bit_data_list
        self.labels_list = labels_list
        
        # Flatten the start indices so PyTorch can just ask for item 'idx'
        self.samples = []
        for array_id, start_idx_array in enumerate(start_indices_list):
            for start_idx in start_idx_array:
                self.samples.append((array_id, start_idx))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        # 1. Look up which CAN ID array to slice, and where to start
        array_id, start_idx = self.samples[idx]
        
        # 2. Slice the 200 frames exactly when needed (0 RAM overhead)
        window_data = self.bit_data_list[array_id][start_idx : start_idx + self.window_size]
        base_labels = self.labels_list[array_id]
        
        # 3. Dynamic Masking calculated on-the-fly!
        # ptp (peak-to-peak) finds the difference between max and min. If 0, it's static.
        static_mask = np.ptp(window_data, axis=0) == 0 
        target_y = base_labels.copy()
        target_y[static_mask] = 0
        
        return torch.tensor(window_data, dtype=torch.float32), torch.tensor(target_y, dtype=torch.long)


def get_labeled_dataloaders(file_path, labels_dict, train_ids, val_ids, test_ids, window_size=200, window_stride=10, batch_size=256):
    print("Loading raw CSV...")
    df = pd.read_csv(file_path, sep='\s+', header=None, names=['timestamp', 'bus', 'raw_frame'])
    
    def hex_to_bits(h):
        return [int(b) for b in bin(int(h, 16))[2:].zfill(64)]

    def process_ids(id_list):
        bit_data_list, labels_list, start_indices_list = [], [], []
        
        for target_id in id_list:
            if target_id not in labels_dict: continue
            
            # Filter
            df_target = df[df['raw_frame'].str.contains(f"^{target_id}#")]
            if len(df_target) < window_size: continue
            
            # Extract
            payloads = df_target['raw_frame'].str.split('#').str[1]
            # Convert directly to int8 array
            bit_data = np.array(list(payloads.apply(hex_to_bits)), dtype=np.int8) 
            
            # Calculate valid starting blocks (0, 10, 20...)
            max_start = bit_data.shape[0] - window_size
            start_indices = np.arange(0, max_start + 1, window_stride)
            
            # Store the raw, lightweight data
            bit_data_list.append(bit_data)
            labels_list.append(np.array(labels_dict[target_id], dtype=np.int8))
            start_indices_list.append(start_indices)
            
        if not bit_data_list: return None
        return CANClassificationDataset(bit_data_list, labels_list, start_indices_list, window_size)

    print("Building Train Dataset...")
    train_ds = process_ids(train_ids)
    print("Building Validation Dataset...")
    val_ds = process_ids(val_ids)
    print("Building Test Dataset...")
    test_ds = process_ids(test_ids)
    
    train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True) if train_ds else None
    val_dl = DataLoader(val_ds, batch_size=batch_size, shuffle=False) if val_ds else None
    test_dl = DataLoader(test_ds, batch_size=batch_size, shuffle=False) if test_ds else None
    
    return train_dl, val_dl, test_dl
```

```python
import torch
import torch.nn as nn
import math

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super().__init__()
        # Create a matrix of positional encodings using sine and cosine waves
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        
        # Shape: (1, max_len, d_model) to broadcast across batches
        self.pe = pe.unsqueeze(0) 

    def forward(self, x):
        # Add the positional encoding to the input embeddings
        return x + self.pe[:, :x.size(1), :].to(x.device)

class CANTransformerClassifier(nn.Module):
    def __init__(self, input_size=64, d_model=128, nhead=4, num_layers=3, 
                 dim_feedforward=256, num_classes=5, dropout=0.2):
        super(CANTransformerClassifier, self).__init__()
        
        self.num_classes = num_classes

        # 1. Input Projection: Expand 64 bits into the wider d_model space
        self.input_projection = nn.Linear(input_size, d_model)
        
        # # 2. Positional Encoding
        # self.pos_encoder = PositionalEncoding(d_model)

        # 3. The Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=dim_feedforward,
            dropout=dropout,
            batch_first=True # Crucial: tells PyTorch our shape is (Batch, Seq, Feature)
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # 4. Final Classification Head
        self.fc = nn.Linear(d_model, 64 * num_classes)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        batch_size = x.size(0)

        # Project and add position data
        x = self.input_projection(x)  # Shape: (batch, 200, d_model)
        # x = self.pos_encoder(x)       # Shape: (batch, 200, d_model)

        # Pass through attention layers
        x = self.transformer_encoder(x) # Shape: (batch, 200, d_model)

        # Sequence Pooling: Average the "understanding" across all 200 time steps
        x = x.mean(dim=1)             # Shape: (batch, d_model)

        # Decode to our final predictions
        x = self.dropout(x)
        out = self.fc(x)              # Shape: (batch, 64 * num_classes)

        # Reshape for CrossEntropyLoss
        out = out.view(batch_size, 64, self.num_classes)

        return out

def Transformer_baseline(dropout=0.2, lr=0.001, d_model=128, nhead=4, 
                         num_layers=3, dim_feedforward=256, num_classes=5):
    """Initializes the Transformer model, loss, and optimizer."""
    model = CANTransformerClassifier(
        input_size=64,
        d_model=d_model,
        nhead=nhead,
        num_layers=num_layers,
        dim_feedforward=dim_feedforward,
        num_classes=num_classes,
        dropout=dropout
    )
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    return model, criterion, optimizer
```

```python
def universal_training_loop(model_initializer, train_dl, val_dl, epochs=50, 
                            device=None, verbose=True, **model_kwargs):
    
    if device is None:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    if verbose:
        print(f"Using device: {device}")

    # Initialize whichever model you passed in (LSTM or Transformer)
    model, criterion, optimizer = model_initializer(**model_kwargs)
    model = model.to(device)

    train_losses, val_losses, train_accs, val_accs = [], [], [], []

    for epoch in range(epochs):
        # Training Phase
        model.train()
        train_loss, train_correct, train_total = 0.0, 0, 0

        for data, target in train_dl:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            
            outputs = model(data)
            
            # Reshape for loss calculation
            outputs = outputs.view(-1, model_kwargs.get('num_classes', 5))
            target = target.view(-1)

            loss = criterion(outputs, target)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            train_total += target.size(0)
            train_correct += (predicted == target).sum().item()

        avg_train_loss = train_loss / len(train_dl)
        train_acc = 100. * train_correct / train_total

        # Validation Phase
        model.eval()
        val_loss, val_correct, val_total = 0.0, 0, 0

        with torch.no_grad():
            for data, target in val_dl:
                data, target = data.to(device), target.to(device)
                
                outputs = model(data)
                outputs = outputs.view(-1, model_kwargs.get('num_classes', 5))
                target = target.view(-1)

                loss = criterion(outputs, target)
                val_loss += loss.item()
                
                _, predicted = torch.max(outputs.data, 1)
                val_total += target.size(0)
                val_correct += (predicted == target).sum().item()

        avg_val_loss = val_loss / len(val_dl)
        val_acc = 100. * val_correct / val_total

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)

        if verbose and (epoch + 1) % 2 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.2f}%, Val Loss: {avg_val_loss:.4f}, Val Acc: {val_acc:.2f}%')

    return model, train_losses, val_losses, train_accs, val_accs
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.patches import Patch
import torch

def plot_labeled_window(window_tensor, bit_labels, can_id, prediction_labels):
    """
    Plots a CAN bit heatmap, a ground truth label strip, and a prediction label strip.
    """
    # 1. Prepare Main Window Data
    data = window_tensor.numpy() if hasattr(window_tensor, 'numpy') else np.array(window_tensor)
    time_steps, bits = data.shape 
    
    # 2. Setup figure with 3 Subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 11), 
                                        gridspec_kw={'height_ratios': [10, 1, 1]},
                                        sharex=True)
    
    # 3. Plot the Main Bit Sequence (The Window)
    ax1.imshow(data, aspect='auto', cmap='binary', interpolation='nearest')
    ax1.set_title(f"Temporal Bit Sequence: ID {can_id}", fontsize=16, pad=20)
    ax1.set_ylabel(f"Time Steps (0-{time_steps})", fontsize=12)
    
    # Define colors for both label strips
    class_colors = ['#f2f2f2', '#e74c3c', '#2ecc71', '#3498db', '#f39c12']
    cmap_labels = mcolors.ListedColormap(class_colors)

    # 4. Plot the Ground Truth Label Strip (ax2)
    label_strip = np.array(bit_labels).reshape(1, -1)
    ax2.imshow(label_strip, aspect='auto', cmap=cmap_labels, vmin=0, vmax=4)
    ax2.set_yticks([]) 
    ax2.set_ylabel("Ground\nTruth", fontsize=10, rotation=0, labelpad=25, va='center')

    # 5. Plot the Prediction Label Strip (ax3) <-- THIS WAS MISSING
    pred_strip = np.array(prediction_labels).reshape(1, -1)
    ax3.imshow(pred_strip, aspect='auto', cmap=cmap_labels, vmin=0, vmax=4)
    ax3.set_yticks([]) 
    ax3.set_xlabel("Bit Index (0-63)", fontsize=12) 
    ax3.set_ylabel("Model\nPredictions", fontsize=10, rotation=0, labelpad=25, va='center')
    
    # 6. Align axes by adding invisible colorbars to match widths
    div1 = make_axes_locatable(ax1)
    cax1 = div1.append_axes("right", size="2%", pad=0.1)
    cax1.axis('off')

    div2 = make_axes_locatable(ax2)
    cax2 = div2.append_axes("right", size="2%", pad=0.1)
    cax2.axis('off') 

    div3 = make_axes_locatable(ax3)
    cax3 = div3.append_axes("right", size="2%", pad=0.1)
    cax3.axis('off')
    
    # 7. Legend
    legend_elements = [
        Patch(facecolor=class_colors[0], label='0: Spare/Static'),
        Patch(facecolor=class_colors[1], label='1: Bit Flag'),
        Patch(facecolor=class_colors[2], label='2: Counter'),
        Patch(facecolor=class_colors[3], label='3: Sensor'),
        Patch(facecolor=class_colors[4], label='4: Checksum'),
    ]
    ax1.legend(handles=legend_elements, loc='lower right')

    # Ensure bits align properly to grid
    ax3.set_xlim(-0.5, 63.5)
    ax3.set_xticks(range(0, 64, 4))
    
    plt.tight_layout()
    plt.show()
```
