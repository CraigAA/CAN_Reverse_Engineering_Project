```python
import cv2
import numpy as np
import torch
import time

print("=== OpenCV Test ===")
print(f"OpenCV version: {cv2.__version__}")

# Test basic OpenCV functionality
print("\n1. Basic OpenCV functionality:")
try:
    # Create a test image
    img = np.zeros((512, 512, 3), dtype=np.uint8)
    cv2.circle(img, (256, 256), 100, (0, 255, 0), -1)
    print("✅ Image creation and drawing: SUCCESS")
except Exception as e:
    print(f"❌ Basic OpenCV failed: {e}")

# Test OpenCV CUDA support
print("\n2. OpenCV CUDA support:")
try:
    cuda_devices = cv2.cuda.getCudaEnabledDeviceCount()
    print(f"CUDA-enabled devices: {cuda_devices}")
    if cuda_devices > 0:
        print("✅ OpenCV compiled with CUDA support")
    else:
        print("⚠️ OpenCV not compiled with CUDA support")
except:
    print("⚠️ OpenCV CUDA functions not available")

# Test image processing performance (CPU vs GPU if available)
print("\n3. Performance test (Gaussian Blur):")
# Create a large test image
test_img = np.random.randint(0, 255, (2000, 2000, 3), dtype=np.uint8)

# CPU processing
start_time = time.time()
cpu_result = cv2.GaussianBlur(test_img, (15, 15), 0)
cpu_time = time.time() - start_time
print(f"CPU processing time: {cpu_time:.4f} seconds")

# Try GPU processing if available
try:
    gpu_img = cv2.cuda_GpuMat()
    gpu_img.upload(test_img)
    gpu_result = cv2.cuda_GpuMat()
    
    start_time = time.time()
    cv2.cuda.bilateralFilter(gpu_img, gpu_result, -1, 50, 50)
    result_cpu = gpu_result.download()
    gpu_time = time.time() - start_time
    
    print(f"GPU processing time: {gpu_time:.4f} seconds")
    print(f"Speedup: {cpu_time/gpu_time:.2f}x")
    print("✅ GPU-accelerated OpenCV working!")
except Exception as e:
    print(f"⚠️ GPU processing not available: {e}")

# Test integration with PyTorch
print("\n4. OpenCV + PyTorch integration:")
try:
    # Convert OpenCV image to PyTorch tensor
    opencv_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
    torch_tensor = torch.from_numpy(opencv_img).permute(2, 0, 1).float() / 255.0
    
    # Move to GPU if available
    if torch.cuda.is_available():
        torch_tensor = torch_tensor.cuda()
        print(f"✅ Image on GPU: {torch_tensor.device}")
        print(f"Tensor shape: {torch_tensor.shape}")
        print(f"Tensor dtype: {torch_tensor.dtype}")
    
    # Convert back to OpenCV format
    if torch_tensor.is_cuda:
        numpy_img = (torch_tensor.permute(1, 2, 0).cpu().numpy() * 255).astype(np.uint8)
    else:
        numpy_img = (torch_tensor.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
    
    opencv_result = cv2.cvtColor(numpy_img, cv2.COLOR_RGB2BGR)
    print("✅ OpenCV ↔ PyTorch conversion: SUCCESS")
    
except Exception as e:
    print(f"❌ OpenCV + PyTorch integration failed: {e}")

print("\n=== Test Complete ===")
```
